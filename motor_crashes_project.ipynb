```python
import pandas as pd
raw_data= pd.read_csv('C:\\Users\\USER\Desktop\\portfolio_datasets\\Motor_Vehicle_Collisions_Crashes.csv')
raw_data.head(5)
```

```python
#drop rows where crash date,time or collision ID has null values 
dropna_data = raw_data.dropna(subset = ['COLLISION_ID', 'CRASH DATE','CRASH TIME'])
```

```python
#drop undesired columns
drop_column_data= dropna_data.drop(['LATITUDE','LONGITUDE','ZIP CODE','LOCATION'], axis=1)
```

```python
#fill null values 
fillnan_data= drop_column_data.fillna({'BOROUGH':'Unknown','ON STREET NAME':'Unknown street','CROSS STREET NAME':'Unknown street',
       'OFF STREET NAME':'Unknown street','NUMBER OF PERSONS INJURED':0,
       'NUMBER OF PERSONS KILLED':0,'NUMBER OF PEDESTRIANS INJURED':0,
       'NUMBER OF PEDESTRIANS KILLED':0,'NUMBER OF CYCLIST INJURED':0,
       'NUMBER OF CYCLIST KILLED':0,'NUMBER OF MOTORIST INJURED':0,
       'NUMBER OF MOTORIST KILLED':0,'CONTRIBUTING FACTOR VEHICLE 1':'Unspecified',
       'CONTRIBUTING FACTOR VEHICLE 2':'Unspecified','CONTRIBUTING FACTOR VEHICLE 3':'Unspecified',
       'CONTRIBUTING FACTOR VEHICLE 4':'Unspecified','CONTRIBUTING FACTOR VEHICLE 5':'Unspecified',
       'VEHICLE TYPE CODE 1':'Unspecified','VEHICLE TYPE CODE 2':'Unspecified',
       'VEHICLE TYPE CODE 3':'Unspecified','VEHICLE TYPE CODE 4':'Unspecified','VEHICLE TYPE CODE 5':'Unspecified'})
```

```python
#identify unique values in reach column
#check for each column
fillnan_data['VEHICLE TYPE CODE 1'].unique()
```

```python
#filter out rows with ambiguous column values
filt= (fillnan_data['CONTRIBUTING FACTOR VEHICLE 1']=='80') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 1']== '1') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 2']== '80') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 2']== '1') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 3']== '80') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 3']== '1') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 4']== '80') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 4']== '1') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 5']== '80') 
| (fillnan_data['CONTRIBUTING FACTOR VEHICLE 5']== '1')
```

```python
#select desired data from filt using (~)
filtered_data = fillnan_data[~filt]
```

```python
#convert data types to string
filtered_data[['BOROUGH','ON STREET NAME','CROSS STREET NAME','OFF STREET NAME','CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5','VEHICLE TYPE CODE 1','VEHICLE TYPE CODE 2','VEHICLE TYPE CODE 3','VEHICLE TYPE CODE 4','VEHICLE TYPE CODE 5']]=filtered_data[['BOROUGH','ON STREET NAME','CROSS STREET NAME','OFF STREET NAME','CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4','CONTRIBUTING FACTOR VEHICLE 5','VEHICLE TYPE CODE 1','VEHICLE TYPE CODE 2','VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4','VEHICLE TYPE CODE 5']].astype(str)
```

```python
#convert data types to integer
filtered_data[['NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED','NUMBER OF PEDESTRIANS INJURED','NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED','COLLISION_ID']]=filtered_data[['NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED','NUMBER OF PEDESTRIANS INJURED','NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED','COLLISION_ID']].astype(int)
```

```python
#convert crash date data type from string to datetime format
filtered_data['CRASH DATE']= pd.to_datetime(filtered_data['CRASH DATE'])
```

```python
#remove whitespaces in the dataframe
filtered_data['BOROUGH']=filtered_data['BOROUGH'].str.strip()
filtered_data['ON STREET NAME']=filtered_data['ON STREET NAME'].str.strip()
filtered_data['OFF STREET NAME']=filtered_data['OFF STREET NAME'].str.strip()
filtered_data['CROSS STREET NAME']=filtered_data['CROSS STREET NAME'].str.strip()
filtered_data['CONTRIBUTING FACTOR VEHICLE 1']=filtered_data['CONTRIBUTING FACTOR VEHICLE 1'].str.strip()
filtered_data['CONTRIBUTING FACTOR VEHICLE 2']=filtered_data['CONTRIBUTING FACTOR VEHICLE 2'].str.strip()
filtered_data['CONTRIBUTING FACTOR VEHICLE 3']=filtered_data['CONTRIBUTING FACTOR VEHICLE 3'].str.strip()
filtered_data['CONTRIBUTING FACTOR VEHICLE 4']=filtered_data['CONTRIBUTING FACTOR VEHICLE 4'].str.strip()
filtered_data['CONTRIBUTING FACTOR VEHICLE 5']=filtered_data['CONTRIBUTING FACTOR VEHICLE 5'].str.strip()
filtered_data['VEHICLE TYPE CODE 1']=filtered_data['VEHICLE TYPE CODE 1'].str.strip()
filtered_data['VEHICLE TYPE CODE 2']=filtered_data['VEHICLE TYPE CODE 2'].str.strip()
filtered_data['VEHICLE TYPE CODE 3']=filtered_data['VEHICLE TYPE CODE 3'].str.strip()
filtered_data['VEHICLE TYPE CODE 4']=filtered_data['VEHICLE TYPE CODE 4'].str.strip()
filtered_data['VEHICLE TYPE CODE 5']=filtered_data['VEHICLE TYPE CODE 5'].str.strip()
```

```python
#replace values
filtered_data['VEHICLE TYPE CODE 1'].replace({'ï¿½MBU':'MBU', 'Dump':'Dump truck','1':'lime moped',"''lime mope":'lime moped'}, inplace = True)
filtered_data['VEHICLE TYPE CODE 2'].replace({'4 dr sedan':'Sedan'}, inplace= True)
filtered_data['CONTRIBUTING FACTOR VEHICLE 1'].replace({'Ilnes':'Illness',
    'Pedestrian/Bicyclist/Other Pedestrian Error/Confusion':'Pedestrian/Bicyclist Error',
    'Drugs (Illegal)':'Illegal drugs','Aggressive Driving/Road Rage':'Aggressive Driving',
    'Cell Phone (hand-held)':'Using cell phone','Driver Inattention/Distraction':'Driver Inattention'}, inplace= True)
filtered_data['CONTRIBUTING FACTOR VEHICLE 2'].replace({'Cell Phone (hand-held)':'Using cell phone',
    'Reaction to Other Uninvolved Vehicle':'Reacting to Uninvolved Vehicle',
    'Reaction to Uninvolved Vehicle':'Reacting to Uninvolved Vehicle','Aggressive Driving/Road Rage':'Aggressive Driving',
    'Driver Inattention/Distraction':'Driver Inattention','Obstruction/Debris':'Obstruction',
    'Pedestrian/Bicyclist/Other Pedestrian Error/Confusion':'Pedestrian/Bicyclist Error'}, inplace= True)
filtered_data['CONTRIBUTING FACTOR VEHICLE 3'].replace({'Reaction to Other Uninvolved Vehicle':'Reacting to Uninvolved Vehicle',
    'Reaction to Uninvolved Vehicle':'Reacting to Uninvolved Vehicle','Aggressive Driving/Road Rage':'Aggressive Driving',
    'Driver Inattention/Distraction':'Driver Inattention','Obstruction/Debris':'Obstruction',
    'Pedestrian/Bicyclist/Other Pedestrian Error/Confusion':'Pedestrian/Bicyclist Error'}, inplace= True)
filtered_data['CONTRIBUTING FACTOR VEHICLE 4'].replace({'Reaction to Uninvolved Vehicle':'Reacting to Uninvolved Vehicle','Aggressive Driving/Road Rage':'Aggressive Driving',
    'Driver Inattention/Distraction':'Driver Inattention','Obstruction/Debris':'Obstruction','Drugs (Illegal)':'Illegal drugs'}, inplace= True)
filtered_data['CONTRIBUTING FACTOR VEHICLE 5'].replace({'Reaction to Uninvolved Vehicle':'Reacting to Uninvolved Vehicle','Aggressive Driving/Road Rage':'Aggressive Driving',
    'Driver Inattention/Distraction':'Driver Inattention','Obstruction/Debris':'Obstruction','Drugs (Illegal)':'Illegal drugs'}, inplace= True)
filtered_data['ON STREET NAME'].replace({'':'Unknown street'}, inplace= True)
filtered_data['OFF STREET NAME'].replace({'':'Unknown street'}, inplace= True)
filtered_data['CROSS STREET NAME'].replace({'':'Unknown street'}, inplace= True)
```

```python
#make column values upper cased
filtered_data['BOROUGH']=filtered_data['BOROUGH'].str.upper()
filtered_data['ON STREET NAME']=filtered_data['ON STREET NAME'].str.upper()
filtered_data['OFF STREET NAME']=filtered_data['OFF STREET NAME'].str.upper()
filtered_data['CROSS STREET NAME']=filtered_data['CROSS STREET NAME'].str.upper()
```

```python
#create day,month and year column from crash date
filtered_data['DAY'] = filtered_data['CRASH DATE'].dt.day_of_week
filtered_data['MONTH'] = filtered_data['CRASH DATE'].dt.month
filtered_data['YEAR'] = filtered_data['CRASH DATE'].dt.year
```

```python
#create day and month name 
filtered_data['DAY'] = filtered_data['DAY'].replace({0:'Monday', 1:'Tuesday',
    2:'Wednesday' ,3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'})
filtered_data['MONTH']= filtered_data['MONTH'].replace({1:'January', 2:'February',3:'March',
    4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
```

```python
#reorder columns
new_order = ['CRASH DATE', 'DAY', 'MONTH', 'YEAR', 'CRASH TIME', 'BOROUGH', 'ON STREET NAME',
       'CROSS STREET NAME', 'OFF STREET NAME', 'NUMBER OF PERSONS INJURED',
       'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
       'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
       'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
       'NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 1',
       'CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3',
       'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
       'COLLISION_ID', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2',
       'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5']
ordered_column_data = filtered_data[new_order]
```
